The code should show the use of the standard library to plot the Gauss and Lorentz random functions.

````python
# import library
import matplotlib.pyplot as plt
import seaborn as sns
from numpy.random import normal as gauss, standard_cauchy as lorentz

#get random value
g=gauss(size=10000)
l=lorentz(100000);

# cut in shot range
g = g[(g>-10) & (g<10)]
l = l[(l>-10) & (l<10)]

#plotting
sns.distplot(g, hist=False, label="Gauss")
sns.distplot(l, hist=False, label="Lorentz")
plt.show()
````
